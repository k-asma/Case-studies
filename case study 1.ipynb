```python
import pandas as pd                     
import matplotlib.pyplot as plt
import seaborn as sns            #importing packages
```

# 1. Reading the data set to python environment

```python
data=pd.read_csv("cars_data.csv")
```

```python
data
```

# 2. Checking for the null value

```python
data.isnull().sum()
```

# 3. Bar graph of male vs female buyers participated in the sales.

```python
data['Buyer Gender'].value_counts().plot(kind='bar', figsize=(7,7),rot=0)
plt.xlabel("Gender")
plt.ylabel("Count of Buyers")
plt.title("Female vs Male Buyers participated in sales")
```

# 4. Top 5 cars based on their sales price.

```python
x=data.sort_values(by=['Sale Price'],ascending=False).head()   #sorting data based on selling price
x.loc[: ,['Make','Model','Sale Price']]  #filtering data
```

# 5. Least 5 cars based on their Resell price.

```python
y=data.sort_values(by=['Resell Price'],ascending=True).head() 
y.loc[: ,['Make','Model','Resell Price']]
```
