## . Gender

```python
df[['gender', 'math score', 'reading score', 'writing score']].groupby('gender').std()
```

Males have comparatively high marks for maths.

```python
sns.boxplot(x="gender", y="writing score", data=df)
```

There is slight difference between the scores of both gender in writing.
